```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,2 * np.pi, 100)
```

```python
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
```

```python
plt.plot(x, y, color='red')
plt.fill(x, y, 'red', alpha=0.4)
plt.title('Heart Shaped Graph')
plt.show()
```
